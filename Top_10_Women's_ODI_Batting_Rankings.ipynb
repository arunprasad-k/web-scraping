# python program to scrape Top 10 ODI batsman in womensâ€™s cricket

```python
#importing required libraries
from bs4 import BeautifulSoup
import requests
import re
```

# sending get request to the webpage server

```python
page=requests.get("https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting")
page
```

# Downloading the webpage source code

```python
page.content
```

# Parse the source code using BeautifulSoup

```python
soup=BeautifulSoup(page.content,"html.parser")
soup
```

# Extracting position tag

```python
position1=soup.find_all("td",class_="rankings-block__position")
position1
```

```python
position2=soup.find_all("td",class_="table-body__cell table-body__cell--position u-text-right")
position2
```

# Extracting texts from the position tag using for loop

```python
pos1=[]
for i in position1:
    pos1.append(i.text.replace("\n",""))
pos1 
```

```python
pos2=[]
for i in position2:
    pos2.append(i.text.replace("\n",""))
pos2  
```

```python
pos2
```

```python
position=pos1+pos2
position
```

```python
pos3=[]
for i in position:
    pos3.append(i.replace("This player has moved down in the rankings since the previous rankings update",""))
pos3  
```

```python
position=[]
for i in pos3:
    position.append(i.replace("This player has moved up in the rankings since the previous rankings update",""))
position  
```

# Extracting player name tag

```python
name1=soup.find_all("div",class_="rankings-block__banner--name-large")
name1
```

```python
name2=soup.find_all("td",class_="table-body__cell rankings-table__name name")
name2
```

# Extracting texts from the players tag using for loop

```python
player1=[]
for i in name1:
    player1.append(i.text)
player1 
```

```python
player2=[]
for i in name2:
    player2.append(i.text.replace("\n",""))
player2
```

```python
players=player1+player2
players
```

```python
players=player1+player2
players
```

# Extracting country tag

```python
team1=soup.find_all("div",class_="rankings-block__banner--nationality")
team1
```

```python
team2=soup.find_all("span",class_="table-body__logo-text")
team2
```

# Extracting texts from the team tag using for loop

```python
country1=[]
for i in team1:
    country1.append(i.text.replace("\n",""))
country1  
```

```python
country2=[]
for i in team2:
    country2.append(i.text.replace("\n",""))
country2  
```

```python
country=country1+country2
country
```

# Extarcting rating tag

```python
rate1=soup.find_all("div",class_="rankings-block__banner--rating")
rate1
```

```python
rate2=soup.find_all("td",class_="table-body__cell rating")
rate2
```

# Extracting texts from the rating tag using for loop

```python
rating1=[]
for i in rate1:
    rating1.append(i.text.replace("\n",""))
rating1  
```

```python
rating2=[]
for i in rate2:
    rating2.append(i.text.replace("\n",""))
rating2  
```

```python
rating=rating1+rating2
rating
```

# checking length of the attributes

```python
print(len(position),len(players),len(country),len(rating))
```
