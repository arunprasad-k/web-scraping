# python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
#importing required libraries
from bs4 import BeautifulSoup
import requests
```

# sending get request to the webpage server

```python
page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

# Downloading the webpage source code

```python
page.content
```

# Parse the source code using BeautifulSoup

```python
soup=BeautifulSoup(page.content,"html.parser")
soup
```

# extracting all the tags having the mw-headline

```python
header=soup.find_all("span",class_="mw-headline")
header
```

# Extracting texts from the tag using for loop

```python
wiki_header=[]
for i in header:
    wiki_header.append(i.text)
wiki_header    
```

# framing the header tag from the source web page

```python
import pandas as pd
wiki_main=pd.DataFrame({})
wiki_main["Heading"]=wiki_header
wiki_main
```
