# python program to scrap book name, author name, genre and book review

```python
#importing required libraries
from bs4 import BeautifulSoup
import requests
import re
```

# sending get request to the webpage server

```python
page=requests.get("https://bookpage.com/reviews")
page
```

# Downloading the webpage source code

```python
page.content
```

# Parse the source code using BeautifulSoup

```python
soup=BeautifulSoup(page.content,"html.parser")
soup
```

# Extarcting book title tag

```python
title=soup.find_all("a",href=re.compile("/reviews/"))
title
```

# Extracting texts from the movie title tag using for loop

```python
book_tit=[]
for i in title:
    book_tit.append(i.text)
book_tit
```

# removing duplicates

```python
book_tit = list(dict.fromkeys(book_tit))
print(book_tit)
```

```python
book_tit.remove('')
book_tit
```

```python
book_tit.remove('Read the Review ')
book_tit
```

# Extarcting author name tag

```python
author=soup.find_all("p",class_="sans bold")
author
```

# Extracting texts from author name tag using for loop

```python
book_author=[]
for i in author:
    book_author.append(i.text.replace("\n",""))
book_author
```

# Extarcting genre tag

```python
genre=soup.find_all("p",class_="genre-links hidden-phone")
genre
```

# Extracting texts from genre tag using for loop

```python
book_genre=[]
for i in genre:
    book_genre.append(i.text.replace("\n",""))
book_genre
```

# Extarcting review tag

```python
review=soup.find_all("p",class_="excerpt")
review
```

# Extracting texts from REVIEW tag using for loop

```python
book_review=[]
for i in review:
    book_review.append(i.text.replace("\n",""))
book_review
```

# checking for equal no of rows

```python
print(len(book_tit),len(book_author),len(book_review),len(book_genre))
```

# creating dataframe using pandas

```python
import pandas as pd
book_details=pd.DataFrame({})
book_details["book name"]=book_tit
book_details["genre"]=book_genre
book_details["author"]=book_author
book_details["review"]=book_review
book_details[0:5]
```
